```python
# Finds freq of pairs of adjacent characters
def freq2(str):
  dict = {}
  duos_big = []

  ran = int(len(str) / 2)

  for i in range(ran):
    duo = str[2*i:2*i+2] # gets i and i+1
    if duo in dict:
      dict[duo] = dict[duo] + 1
    else:
      dict[duo] = 1

  duos = []
  for i in dict.keys():
    duos.append(i)
  duos.sort()

  frequency = []
  for i in dict.keys():
    if dict[i] > 10:
      duos_big.append(i)
      frequency.append(dict[i])

  plt.figure(figsize=(10,10))

  plt.bar(duos_big, frequency, 0.8)
  plt.show()
```

```python
# turn a string into pairs of letters according to playfair rules
def alterPlainText(plaintext):
    altplaintext = []
    textlength = len(plaintext)
    j = 0
    for i in range(len(plaintext)):
        if (j >= textlength):
            break
        if (j + 1 == textlength):
            #print([plaintext[j], 'X']) # testing lining
            altplaintext.append([plaintext[j], 'X'])
            break
        if (plaintext[j] == plaintext[j+1]):
            #print([plaintext[j], 'X']) # testing lining
            altplaintext.append([plaintext[j], 'X'])
            j = j+1
        else:
            #print([plaintext[j], plaintext[j+1]]) # testing lining
            altplaintext.append([plaintext[j], plaintext[j+1]])
            j = j + 2
    return altplaintext
```

```python
# playfair encryption
def playfair(matrix, plaintext):
  ciphertext = []
  altplaintext = alterPlainText(plaintext)
  for pair in altplaintext:
    l0_loc = np.where(matrix == pair[0])
    l1_loc = np.where(matrix == pair[1])
    # adjust to [row, col]
    l0_loc = [l0_loc[0][0], l0_loc[1][0]]
    l1_loc = [l1_loc[0][0], l1_loc[1][0]]
    # same row?
    if(l0_loc[0] == l1_loc[0]):
      #print("same row")
      firstletter = matrix[l0_loc[0]][(l0_loc[1] + 1) % 5]
      secondletter = matrix[l1_loc[0]][(l1_loc[1] + 1) % 5]
      ciphertext.append([firstletter, secondletter])
    elif(l0_loc[1] == l1_loc[1]):
      #print("same col")
      firstletter = matrix[(l0_loc[0] + 1) % 5][l0_loc[1]]
      secondletter = matrix[(l1_loc[0] + 1) % 5][l0_loc[1]]
      ciphertext.append([firstletter, secondletter])
    # otherwise its a rectangle
    else:
      # rectangle, its the same row, but a diffent column 
      #print("rectangle")
      firstletter = matrix[l0_loc[0]][l1_loc[1]]
      secondletter = matrix[l1_loc[0]][l0_loc[1]]
      ciphertext.append([firstletter, secondletter])
  return ciphertext
    
```

```python
# Playfair Decryption
def decrypt(keyMatrix, ciphertext):
    numPairs = int(len(ciphertext) / 2)
    plaintext = ''

    # Take a digraph and decrypt it
    for i in range(numPairs):
        pair = ciphertext[i * 2] + ciphertext[i * 2 + 1]

        loc1 = np.where(keyMatrix == pair[0])
        loc2 = np.where(keyMatrix == pair[1])
        # adjust to [row, col]
        loc1 = [loc1[0][0], loc1[1][0]]
        loc2 = [loc2[0][0], loc2[1][0]]

        # If letters are in the same column
        if loc1[1] == loc2[1]:
            result = shiftCol(keyMatrix, pair, loc1, loc2)
            plaintext = plaintext + result

        # If letters are in the same row
        elif loc1[0] == loc2[0]:
            result = shiftRow(keyMatrix, pair, loc1, loc2)
            plaintext = plaintext + result
        
        # If letters are in a rectangle
        else :
            result = shiftRec(keyMatrix, pair, loc1, loc2)
            plaintext = plaintext + result

    return plaintext
```

```python
# Decrypt if letters are in same column (Shift up 1)
def shiftCol(matrix, digraph, loc1, loc2):
    letter1 = digraph[0]
    letter2 = digraph[1]

    if loc1[0] == 0:
        result1 = matrix[4][loc1[1]]
    else:
        result1 = matrix[loc1[0] - 1][loc1[1]]

    if loc2[0] == 0:
        result2 = matrix[4][loc2[1]]
    else:
        result2 = matrix[loc2[0] - 1][loc2[1]]

    return result1 + result2
```

```python
# Decrypt if letters are in same row (Shift left 1)
def shiftRow(matrix, digraph, loc1, loc2):
    letter1 = digraph[0]
    letter2 = digraph[1]

    if loc1[1] == 0:
        result1 = matrix[loc1[0]][4]
    else:
        result1 = matrix[loc1[0]][loc1[1] - 1]

    if loc2[1] == 0:
        result2 = matrix[loc2[0]][4]
    else:
        result2 = matrix[loc2[0]][loc2[1] - 1]

    return result1 + result2
```

```python
# Decrypt if letters are in a rectangle (Swap corners)
def shiftRec(matrix, digraph, loc1, loc2):
    result1 = matrix[loc1[0]][loc2[1]]
    result2 = matrix[loc2[0]][loc1[1]]

    return result1 + result2
```
